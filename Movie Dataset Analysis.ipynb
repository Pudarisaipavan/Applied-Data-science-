```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Downloads/imdb_movie_dataset.csv')
df
```

```python
# Calculate correlation matrix for numerical columns
correlation_matrix = df.corr()
print(correlation_matrix)
```

```python
def plot_movies_per_year(data):
    # Count movies per year and sort by year
    movies_per_year = data['Year'].value_counts().sort_index()

    plt.figure(figsize=(10, 6))
    plt.plot(movies_per_year.index, movies_per_year.values, color='g', marker='o', linestyle='-')
    
    plt.xlabel('Year')
    plt.ylabel('Number of Movies Released')
    plt.title('Number of Movies Released by Year')

    # Annotate the points with the movie counts
    for x, y in zip(movies_per_year.index, movies_per_year.values):
        plt.text(x, y + 0.15, str(y), ha='center', va='bottom', fontsize=9)

    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
plot_movies_per_year(df)
```

```python
def plot_top_genres(data):
    # Split the genres and explode them to get a count of each genre
    genre_counts = data['Genre'].str.split(',').explode().value_counts().head(10)
    
    plt.figure(figsize=(10, 6))
    genre_counts.plot(kind='bar', color='skyblue')
    
    plt.xlabel('Genre')
    plt.ylabel('Number of Movies')
    plt.title('Top 10 Most Filmed Genres')
    
    # Annotate each bar with the count value
    for index, value in enumerate(genre_counts):
        plt.text(index, value + 5, str(value), ha='center', va='bottom', fontsize=9)
    
    plt.xticks(rotation=45)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
plot_top_genres(df)
```

```python
def plot_rating_vs_revenue(data):
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x='Rating', y='Revenue (Millions)', data=data, alpha=0.7, color='coral')
    plt.title('IMDb Rating vs. Revenue')
    plt.xlabel('IMDb Rating')
    plt.ylabel('Revenue (Millions)')
    plt.grid(axis='both', linestyle='--', alpha=0.7)
    plt.show()
plot_rating_vs_revenue(df)
```
